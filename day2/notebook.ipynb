*(In your input, these appear on a single long line.)*

### Rules  
- Ranges are separated by commas `,`.  
- Each range gives its first and last ID separated by a dash `-`.  
- An **invalid ID** is any number made of a sequence of digits repeated twice:  
    - Examples: `55` (5 twice), `6464` (64 twice), `123123` (123 twice).  
- No leading zeroes allowed: `0101` is invalid as an ID, but `101` is valid.  

Your job: **find all invalid IDs in the given ranges**.

---

### Example Analysis  
For the example above:  

- `11-22` â†’ invalid IDs: `11`, `22`  
- `95-115` â†’ invalid ID: `99`  
- `998-1012` â†’ invalid ID: `1010`  
- `1188511880-1188511890` â†’ invalid ID: `1188511885`  
- `222220-222224` â†’ invalid ID: `222222`  
- `1698522-1698528` â†’ none  
- `446443-446449` â†’ invalid ID: `446446`  
- `38593856-38593862` â†’ invalid ID: `38593859`  
- Others â†’ none  

Adding them up:  

11 + 22 + 99 + 1010 + 1188511885 + 222222 + 446446 + 38593859 = 1227775554

### Puzzle Question  
**What do you get if you add up all of the invalid IDs in your input?**

````python
with open('input_full.txt', 'r') as f: # -> part1: 23701357374, part 2: 
# with open("input_full.txt", 'r') as f: # -> part1: 1227775554, part 2:
    input = f.read().splitlines()
ranges = input[0].split(",")


def is_double_repeat(s: str) -> bool:
    """
    Return True iff s is exactly two copies of the same block.
    Examples:
      '55' -> True           (block '5' repeated twice)
      '6464' -> True         (block '64' repeated twice)
      '111' -> False         (length is odd, cannot be two equal halves)
      'abababab' -> True     (two copies of 'abab')
      'abcab' -> False       (length odd)
      '' -> False            (empty string: by convention return False)
    """
    n = len(s)
    if n % 2 != 0 or n == 0:
        return False
    return s[:n // 2] == s[n // 2:]


counter = 0
for range_ in ranges:
    first, second = range_.split("-")[0], range_.split("-")[1]
    for i in range(int(first), int(second) + 1):
        if is_double_repeat(str(i)):
            counter += i
counter
````

## --- Part Two ---

The clerk quickly discovers that there are still invalid IDs in the ranges in your list. Maybe the young Elf was doing other silly patterns as well?

Now, an ID is invalid if it is made only of some sequence of digits repeated at least twice. So, 12341234 (1234 two times), 123123123 (123 three times), 1212121212 (12 five times), and 1111111 (1 seven times) are all invalid IDs.

From the same example as before:

```
    11-22 still has two invalid IDs, 11 and 22.
    95-115 now has two invalid IDs, 99 and 111.
    998-1012 now has two invalid IDs, 999 and 1010.
    1188511880-1188511890 still has one invalid ID, 1188511885.
    222220-222224 still has one invalid ID, 222222.
    1698522-1698528 still contains no invalid IDs.
    446443-446449 still has one invalid ID, 446446.
    38593856-38593862 still has one invalid ID, 38593859.
    565653-565659 now has one invalid ID, 565656.
    824824821-824824827 now has one invalid ID, 824824824.
    2121212118-2121212124 now has one invalid ID, 2121212121.
```

Adding up all the invalid IDs in this example produces 4174379265.

What do you get if you add up all of the invalid IDs using these new rules?

````python
with open('input_full.txt', 'r') as f: # -> part 2: 4174379265
# with open("input_full.txt", 'r') as f: # -> part 2:
    input = f.read().splitlines()
ranges = input[0].split(",")


def is_valid_id(s: str) -> bool:
    """
    Return True if s can be written as t repeated k times for some t and k > 1.
    Examples: '55' -> True, '6464' -> True, 'aaaa' -> True, 'abc' -> False.
    """
    # Empty or single-character strings cannot have k>1 repeating structure
    if len(s) <= 1:
        return False
    result = s in (s + s)[1:-1]
    return result


counter = 0
for range_ in ranges:
    first, second = range_.split("-")[0], range_.split("-")[1]
    for i in range(int(first), int(second) + 1):
        if is_valid_id(str(i)):
            counter += i
counter
````
